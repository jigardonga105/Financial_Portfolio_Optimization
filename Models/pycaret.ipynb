```python
import pycaret
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
pycaret.__version__
```

```python
indexes = []
all_indexes = {}

all_indexes = pd.read_excel('../Data/Index_Data.xlsx', sheet_name=None)

for sheet_name, df in all_indexes.items():
    indexes.append(sheet_name)
    all_indexes[sheet_name] = df

BSESN = all_indexes['BSESN']
BSESN.tail()
```

```python
from pycaret.regression import *
import pandas as pd

df = BSESN[['Open', 'Close', 'Momentum_Indicator', 'High_Low_Change', 'High_Open_Change', 'Low_Open_Change', 'Day', 'Month', 'Year']]

# Initialize PyCaret setup
reg_setup = setup(data=df, 
                  target='Close')

# Compare models to find the best one
best_model = compare_models()

print(best_model)

# Create the best model
model = create_model(best_model)

# Tune the model
tuned_model = tune_model(model)

# Evaluate the model
evaluate_model(tuned_model)

# Finalize the model
final_model = finalize_model(tuned_model)

# Save the model
save_model(final_model, 'final_model')
```
